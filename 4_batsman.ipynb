```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
class BatsmanState(TypedDict):
    runs:int
    balls:int
    fours:int
    sixs:int

    sr:float
    bpb:float
    boundry_persentage:float
    
```

```python
def calculate_sr(State:BatsmanState):
    sr=(State['runs']/State['balls'])*100
    return {'sr':sr}
```

```python
def calculate_bpb(State:BatsmanState):
    bpb=State['balls']/(State['fours'] + State['sixs'])
    return {'bpb':bpb}
```

```python
def calculate_boundry_persentage(State:BatsmanState):
    boundry_persentage=(((State['fours']*4) + (State['sixs']*6))/State['runs'])*100
    return {'boundry_persentage':boundry_persentage}
```

```python
def summary(state:BatsmanState):
    summary=f"""
Strick_rate=\n{state['sr']}\n
bounder par balls=\n{state['bpb']}\n
boundry precentage=\n{state['boundry_persentage']}

"""
```

```python
graph=StateGraph(BatsmanState)

graph.add_node('calculate_sr',calculate_sr)
graph.add_node('calculate_bpb',calculate_bpb)
graph.add_node('calculate_boundry_persentage',calculate_boundry_persentage)
graph.add_node('summary',summary)

graph.add_edge(START,'calculate_sr')
graph.add_edge(START,'calculate_bpb')
graph.add_edge(START,'calculate_boundry_persentage')

graph.add_edge('calculate_sr','summary')
graph.add_edge('calculate_bpb','summary')
graph.add_edge('calculate_boundry_persentage','summary')

graph.add_edge('summary',END)

workflow=graph.compile()
```

```python
initial_state={'runs':216,
    'balls':120,
    'fours':15,
    'sixs':10,
    }

final_state=workflow.invoke(initial_state)
print(final_state)
```
